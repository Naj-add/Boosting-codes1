```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor

# Création du dataset
data = {
    "Surface": [100, 150, 200, 250, 300],
    "Chambres": [2, 3, 4, 4, 5],
    "Localisation": ["A", "B", "A", "C", "B"],
    "Prix": [150000, 200000, 250000, 300000, 350000]
}
df = pd.DataFrame(data)
df
df_encoded = pd.get_dummies(df, columns=["Localisation"])
X = df_encoded.drop("Prix", axis=1)
y = df_encoded["Prix"]

model = GradientBoostingRegressor(
    n_estimators=50,
    learning_rate=0.05,
    max_depth=2,
    random_state=42
)
model.fit(X, y)
predictions = model.predict(X)

resultats = df.copy()
resultats["Prediction"] = predictions
resultats["Erreur (résidu)"] = y - predictions

resultats
```

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor(
    n_estimators=50,
    learning_rate=0.1,
    max_depth=3,
    objective="reg:squarederror",
    random_state=42
)
xgb_model.fit(X, y)
pred_xgb = xgb_model.predict(X)

resultats_xgb = df.copy()
resultats_xgb["Prediction_XGBoost"] = pred_xgb
resultats_xgb["Erreur"] = y - pred_xgb

resultats_xgb
```
